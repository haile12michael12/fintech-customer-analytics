```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from transformers import pipeline
```

```python
df = pd.read_csv("thematic_analysis.csv")
# Sentiment Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='bank')
plt.title("Sentiment Distribution by Bank")
plt.savefig("sentiment_distribution.png")
plt.show()
```

```python
# Rating Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='rating', multiple='dodge', binwidth=1)
plt.title("Rating Distribution by Bank")
plt.savefig("rating_distribution.png")
plt.show()
```

```python
# Rating Histogram
```

```python
sns.histplot(data=df, x='rating', hue='bank', multiple='dodge', binwidth=1)
plt.title("Rating Distribution by Bank")
plt.savefig("rating_distribution.png")
```

```python
#Word Cloud by Bank (Positive Themes)
```

```python
all_text = " ".join(df['review'].str.lower())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Common Words in Reviews")
plt.tight_layout()
plt.savefig("keyword_wordcloud.png")
plt.close()

print("Visualizations saved.")
```
